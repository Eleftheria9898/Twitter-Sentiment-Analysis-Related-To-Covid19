```python
# find Hashtags

import re

def hashtags(text):
    hashtag_list = re.findall(r"#(\w+)",text)  # find all the hashtags in a tweet 
    hashtags = ' '.join([str(elem) for elem in hashtag_list])  # Hashtags of the tweet 
    return hashtags
```

```python
tweets['#hashtags']=tweets['tweet_text'].apply(hashtags)
tweets
```
